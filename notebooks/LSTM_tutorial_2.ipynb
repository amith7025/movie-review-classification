```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim
```

# **Tutorial**

```python
torch.manual_seed(1)
```

```python
lstm = nn.LSTM(3,3)
```

```python
inputs = [torch.rand(1,3) for _ in range(5)]
print(inputs)
```

```python
hidden = (
    torch.randn(1,1,3),
    torch.randn(1,1,3)
)
hidden
```

```python
inputs[0],inputs[0].view(1,1,-1)
```

```python
for i in inputs:
    out,hidden = lstm(i.view(1,1,-1),hidden)
```

```python
print(out)
print(hidden)
```
